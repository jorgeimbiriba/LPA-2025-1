```python
import pandas as pd

# Substitua pelo caminho completo do seu arquivo dadosmunicipios.xlsx
# Se ele estiver na mesma pasta onde você está executando o código, basta o nome do arquivo.
caminho_do_arquivo_idh_novo = 'dadosmunicipios.xlsx'

try:
    df_idh_novo = pd.read_excel(caminho_do_arquivo_idh_novo)
    print("Colunas disponíveis no arquivo 'dadosmunicipios.xlsx':")
    print(df_idh_novo.columns.tolist())
    print("\nPrimeiras 5 linhas do arquivo 'dadosmunicipios.xlsx' para visualização:")
    print(df_idh_novo.head())

    # Se houver uma coluna 'ANO' (ou similar), vamos contar os registros por ano para ver a cobertura
    if 'ANO' in df_idh_novo.columns:
        print("\nContagem de registros por ano:")
        print(df_idh_novo['ANO'].value_counts().sort_index())
    elif 'Ano' in df_idh_novo.columns: # Tentar 'Ano' também, caso seja maiúscula
        print("\nContagem de registros por ano:")
        print(df_idh_novo['Ano'].value_counts().sort_index())

except FileNotFoundError:
    print(f"Erro: O arquivo '{caminho_do_arquivo_idh_novo}' não foi encontrado.")
    print("Certifique-se de que o caminho e o nome do arquivo estão corretos.")
except Exception as e:
    print(f"Ocorreu um erro ao ler o arquivo Excel: {e}")
    print("Verifique se o arquivo está no formato correto e não está corrompido.")
```
